**8. Write a program that prompts a user to enter two different numbers. Perform basic arithmetic operations based on the choices.**


```
Enter the first number:15
Enter the Second number:10
1) Addition
2) Subtraction
3) Multiplication
4) Division
Please Enter the Choice:3
Multiplication of 15.0 and 10.0 is: 150.0
```



````python
a=int(input("enter the first number"))
b=int(input("enter the second number"))
d=a+b
e=a-b
f=a*b
g=a/b

c=input("enter the choice  ")

if c=='d':
  print("addition of ",d)
if c=='e':
  print("subtraction of",e)
if c=='f':
  print("multiplication of",f)
if c=='g':
  print("division of",g)
````

**10. Write a program to prompt a user to enter a day of the week. If the entered day if the week is between 1 and 7 then display the respective name of the day.**


```
Input
Enter the day of week: 7
Output
Its Sunday
```



````python
a=int(input("enter the day of week:  "))
if a == 1:
  print("it is monday")
if a==2:
  print("it is tuesday")
if a==3:
  print("it is wednesday")
if a==4:
  print("it is thursday")
if a==5:
  print("it is friday")
if a==6:
  print("it is Saturday ")
if a==7:
  print("it is sunday")

````
